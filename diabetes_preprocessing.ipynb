## Shuffle the data

```python
from sklearn.utils import shuffle
```

```python
X = np.array(df.iloc[:,:-1])
```

```python
y = np.array(df.iloc[:,-1])
```

```python
X,y = shuffle(X,y, random_state=42)
X,y
```

```python
columns = df.columns[:-1]
data_shuffled = pd.DataFrame(X, columns=columns)
data_shuffled['Outcome'] = y
data_shuffled
```

```python
type(data_shuffled['Outcome'][0])
```

## Balancing the data

```python
indeces_to_remove = []
counter = 0
for outcomes in range(data_shuffled.shape[0]):
    if data_shuffled['Outcome'].iloc[outcomes] == 0:
        counter += 1
        if counter > 268:
            indeces_to_remove.append(outcomes)
balanced_data = np.delete(data_shuffled, indeces_to_remove, axis = 0)
```

```python
data_shuffled.columns.values
```

```python
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
       'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
```

```python
balanced_df = pd.DataFrame(balanced_data, columns=columns)
```

```python
balanced_df
```

```python
balanced_df.info
```

## Shuffle again

```python
## since excess zeroes are removed, most of the last rows of the data is 1. we have to reshuffle to avoid overfitting
```

```python
X = np.array(balanced_df.iloc[:,:-1])
y = np.array(balanced_df.iloc[:,-1])
```

```python
X,y = shuffle(X,y, random_state=42)
X,y
```

```python
columns = balanced_df.columns[:-1]
diabetes_data = pd.DataFrame(X, columns=columns)
diabetes_data['Outcome'] = y
diabetes_data
```

## checkpoint
